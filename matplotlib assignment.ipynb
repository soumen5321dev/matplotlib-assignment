```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this
# generated data plot a scatter plot.

# Ans: A scatter plot is a diagram where each value in the data set is represented by a dot.

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter plot graph")
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:

# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])    

# Ans: The subplot() function in Matplotlib is used to create multiple plots within a single figure.
# It allows us to divide the figure into a grid of subplots and specify the location of each subplot
# in the grid. This is particularly useful for comparing different datasets or visualizations side by
# side in a structured way.


import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure
plt.figure(figsize=(10, 8))

# Plot 1
plt.subplot(2, 2, 1)  # (rows, columns, subplot number)
plt.plot(x1, y1)
plt.title('Line Plot 1')
plt.xlabel('x')
plt.ylabel('y')

# Plot 2
plt.subplot(2, 2, 2)  # (rows, columns, subplot number)
plt.plot(x2, y2)
plt.title('Line Plot 2')
plt.xlabel('x')
plt.ylabel('y')

# Plot 3
plt.subplot(2, 2, 3)  # (rows, columns, subplot number)
plt.plot(x3, y3)
plt.title('Line Plot 3')
plt.xlabel('x')
plt.ylabel('y')

# Plot 4
plt.subplot(2, 2, 4)  # (rows, columns, subplot number)
plt.plot(x4, y4)
plt.title('Line Plot 4')
plt.xlabel('x')
plt.ylabel('y')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)

# Ans: A box plot, also known as a box-and-whisker plot, is a standardized way of displaying
#      the distribution of data based on a five-number summary: minimum, first quartile (Q1),
#      median, third quartile (Q3), and maximum. It can also highlight outliers in the data.
#      Box plots are particularly useful for comparing distributions between several groups
#      or datasets.
        

# Box plots are used in Matplotlib for several important reasons, primarily related to their
# ability to succinctly and effectively convey key statistical information about a dataset.
# Here are some of the main reasons why box plots are used:


# 1. Visualizing the Distribution of Data
# Box plots provide a clear visualization of the distribution of a dataset. By showing the
# median, quartiles, and potential outliers, box plots offer a comprehensive snapshot of the
# datas spread and central tendency.

# 2. Identifying Outliers
# Box plots are particularly useful for identifying outliers. Data points that fall outside the
# whiskers of the box plot are considered outliers, which can be crucial for understanding anomalies
# or errors in the dataset.

# 3. Comparing Multiple Datasets
# Box plots allow for easy comparison between multiple groups or categories. By placing multiple box
# plots side by side, you can quickly compare the distributions, medians, and variability of different
# datasets.

# 4. Summarizing Data with Minimal Assumptions
# Box plots provide a summary of the data without making any assumptions about the underlying distribution
# (e.g., normality). This makes them a non-parametric tool that is widely applicable to various types of data.

# 5. Identifying Skewness and Spread
# The position of the median line within the box, the length of the whiskers, and the presence of outliers
# can give insights into the skewness and spread of the data. For example, if the median is closer to the
# bottom of the box and the top whisker is longer, the data might be positively skewed.


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create a box plot
plt.boxplot(data, vert=True, patch_artist=True, labels=['Box 1', 'Box 2'])

# Add title and labels
plt.title('Box Plot of Two Datasets')
plt.xlabel('Dataset')
plt.ylabel('Values')

# Display the plot
plt.show()
```
